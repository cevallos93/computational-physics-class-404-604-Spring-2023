```python
"""
This is the FCTS method for solving the diffusion equation.
Inside the python function one has to write the initial conditions and boundary conditions.
The ones concerning the problem are currently active, while those that have to do with 'other experiments'
are commented out.
"""


"""
Note: If we decrease the space step Δx without a simultaneous quadratic increase in the time step,
we will worsen the stability.
"""

Nx = 101 # number of points + boundary
Dt = 0.00001 # delta t (spacing)
tend = 1.
div = 2000


# Nx, Dt, final time, divisor over Nt (which controls the amount of lines in the graph, equally spaced in time)

def FTCS(Nx, Dt, tend, div):

    D=1.
    Nt = tend/Dt # total time steps - going from time t=0 to t=0.1
    div = 2000 # divisor for my graph number
    
    
    Dx = 1./(Nx-1); # delta x spacing; from 0 to 1 in Nx-1 spaces                                                 
    
    print("Δt = {:.2E}".format(Dt))
    print("Δx = {:.2E}".format(Dx))
    print("Nt = {:.2E}".format(Nt))
    
    # STABILITY CONDITION
    if Dt < 0.5*(Dx**2)/D:
        print("Stability guaranteed. You may continue good sir.")
    else:
        raise SystemExit("EEEEEEEEEEEEEEEEE!!!!!!!")
    
    
    U = np.zeros([Nx,2]);
    
    U_pl = np.zeros([Nx,int(Nt/div)+1])
    
    xarr = np.zeros(Nx) # our point in the x-space grid
    
    # INITIAL CONDITION
    #mu = 0.5;
    #sigma = 0.05;
    for ix in range (0, Nx):  
        xarr[ix]=ix*Dx #i times delta x
        U[ix,0] = 0.5*(np.cos(13*xarr[ix])+1) # Initial U
        
        # Little experiment (IGNORE)
        #U[ix,0] = np.exp(-(xarr[ix]-mu)**2/(2*sigma**2))/np.sqrt(2*np.pi*sigma**2) # Initial U
        
        
    # BOUNDARY CONDITIONS
    U[0,0] = 1. #for hw it's 1, for gaussian it's 0
    U[0,1] = 1.
    U[Nx-1,0] = 0.
    U[Nx-1,1] = 0.
    
    # Boundary conditions for little experiment
    #U[0,0] = 0.
    #U[0,1] = 0.
    
    
    cons = D*Dt*Dx**-2 # constant
    m = 0 # counter selected plots to show
    toutn=0 # counter of iterations
    
    clr = ['xkcd:dark red','xkcd:red','xkcd:orange','xkcd:green','xkcd:blue','xkcd:indigo','xkcd:black']*100
    fig = plt.figure(figsize = (10,10))
    ax = fig.add_subplot(111)
    ax.set_ylim(-0.1,1.1)
    ax.set_xlim(0.,1.)
    ax.set_xlabel(r'$x$')
    ax.set_ylabel(r'$U(x,t)$')
    
    
    t=0.
    while t < tend:                                  
        for ix in range (1,Nx-1):            
            U[ix, 1] = U[ix, 0] +  cons*(U[ix+1, 0] + U[ix-1, 0] - 2.*U[ix,0])
            
        if toutn%2000 == 0: # Every 2000 steps
            for ix in range (1,Nx-1):
                U_pl[ix, m] = U[ix, 1]
            
            ax.plot(xarr, U[:,1],color=clr[m],linestyle='-',label=(r'$t={:.2f}$'.format(t)))
            ax.legend(loc='upper right')
            print("m = ",m)
            m = m + 1
            
        for ix in range (1, Nx - 1):
            U[ix, 0] = U[ix, 1]
        
        t+=Dt
        toutn+=1
```

```python
Nx = 101 # number of points + boundary
Dt = 0.00001 # delta t (spacing)
tend = 0.1
div = 2000

FTCS(Nx, Dt, tend, div)
```
